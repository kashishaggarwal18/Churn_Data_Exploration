```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Customer Churn.csv")
```

```python
# replacing Blanks with 0 in total charges column 
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", 0)
```

```python
df["TotalCharges"] = df["TotalCharges"].astype("float")
df["SeniorCitizen"] = df["SeniorCitizen"].astype("int")
```

```python
df.info()
```

```python
df.describe()
```

```python
df["SeniorCitizen"] = df["SeniorCitizen"].replace({0:"No",1: "Yes"})
```

```python
df["SeniorCitizen"]
```

```python
ax = sns.countplot(x = "Churn", data = df)
ax.bar_label(ax.containers[0])
plt.title("No of Churned Customers")
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show()
```

```python
# from above pie chart we concluded that 26% of our customers has churned out
# now let's find the reason behind it
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gende")
ax.bar_label(ax.containers[0])
```

```python
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Senior Citizens ")
```

```python
total_counts = df.groupby("SeniorCitizen")["Churn"].value_counts(normalize = True).unstack()*100
ax=total_counts.plot(kind = "bar", stacked = True)
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(
            p.get_x() + p.get_width()/2,
            p.get_y() + height /2,
            f'{height:.1f}%',
            ha = "center",
            va = "center",
            fontsize = 10,
            color = "White")
plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', loc = "upper right")  # Customize legend location
```

```python
# looking at the chart aboove we can see that senior citizens have churned more as compared to non senior citizens
```

```python
plt.figure(figsize = (9,4))
sns.histplot( x = "tenure", data = df, bins = 72, hue = "Churn")
plt.title("Churn based on tenure")
```

```python
# from the graph above we observed that people who have used our
#services for long time have stayed and people who have used our services for one or two moths have churned
```

```python
sns.countplot(x = "Contract", data = df, hue = "Churn")
```

```python
total_counts = df.groupby("Contract")["Churn"].value_counts(normalize = True).unstack()*100
ax = total_counts.plot(kind = "bar", stacked = True)
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(
            p.get_x() + p.get_width() /2,
            p.get_y() + height /2,
            f'{height:.1f}%',
            ha = "center",
            va = "center",
            fontsize = 10,
            color = "white")
plt.title("Churning by contracts")
plt.xlabel("Contract")
plt.ylabel("Percentage %")
plt.xticks(rotation = False)
plt.legend(title = "churn", loc = "upper right")
```

```python
# from above chart we concluded that people having month-to-month contract have churned the most
```

```python
column = ['PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies']
fig, axes = plt.subplots(3,3, figsize = (15,15))
axes = axes.flatten()
for i, col in enumerate(column):
    sns.countplot(x = col, data = df, ax = axes[i], hue = "Churn")
    axes[i].set_title(col, size = 15)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("count")
plt.tight_layout()
plt.show()
```

```python
# The majority of customers who do not churn tend to have services like PhoneService, InternetService (particularly DSL), and OnlineSecurity enabled. 
# For services like OnlineBackup, TechSupport, and StreamingTV, churn rates are noticeably higher when these services are not used or are unavailable.
```
